```python
import requests
import os

def download_from_url(url, dst=None):
    if dst is None:
        dst = os.path.basename(url)


    # Download the file
    r = requests.get(url, verify=False)
    r.raise_for_status()  # raise error if download failed

    # Save locally
    with open(dst, "wb") as f:
        f.write(r.content)

    print("Download complete")
```
